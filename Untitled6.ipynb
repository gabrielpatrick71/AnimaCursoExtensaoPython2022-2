```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

velocidade = ctrl.Antecedent(np.arange(0, 11, 1), 'velocidade')
marcha = ctrl.Antecedent(np.arange(0, 11, 1), 'marcha')
inclinacaoTerreno = ctrl.Consequent(np.arange(0, 11, 1), 'inclinação do terreno')

velocidade.automf(names=['alta', 'média', 'baixa'])

marcha['baixa'] = fuzz.trimf(marcha.universe, [0, 0, 5])
marcha['média'] = fuzz.gaussmf(marcha.universe, 5, 2)
marcha['alta'] = fuzz.gaussmf(marcha.universe, 10,3)

inclinacaoTerreno['muito baixo'] = fuzz.trimf(inclinacaoTerreno.universe, [0, 0, 5])
inclinacaoTerreno['baixo'] = fuzz.gaussmf(inclinacaoTerreno.universe, 5, 2)
inclinacaoTerreno['médio'] = fuzz.gaussmf(inclinacaoTerreno.universe, 11,1)
inclinacaoTerreno['alto'] = fuzz.gaussmf(inclinacaoTerreno.universe, 10,3)
inclinacaoTerreno['muito alto'] = fuzz.gaussmf(inclinacaoTerreno.universe, 10,3)

velocidade.view()
marcha.view()
inclinacaoTerreno.view()

rule1 = ctrl.Rule(marcha['alta'] | velocidade['alta'], inclinacaoTerreno['baixo'])
rule2 = ctrl.Rule(marcha['média'], inclinacaoTerreno['médio'])
rule3 = ctrl.Rule(marcha['baixa'] & velocidade['baixa'], inclinacaoTerreno['alto'])

inclinacaoTerreno_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
inclinacaoTerreno_simulador = ctrl.ControlSystemSimulation(inclinacaoTerreno_ctrl)
```
